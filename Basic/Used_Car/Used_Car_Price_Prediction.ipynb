# **4. Data Acquisition & Description**
---

This section is emphasised on the accquiring the data and obtain some descriptive information out of it.

This data set consists of three types of entities:
- The specification of an auto in terms of various characteristics
- The assigned insurance risk rating
- Normalized losses in use as compared to other cars

The first factor is the generic characteristics of the specifications.  

The second factor rating corresponds to the degree to which the auto is riskier than its price indicates.

- Cars are initially assigned a risk factor symbol associated with its price.
- Then, if it is more or less risky this symbol is adjusted by moving it up or down the scale.
- Actuarius calls this process "symbolling". A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe.

The third factor is the relative average loss payment per insured vehicle year.

This value is normalized for all autos within a particular size classification (two-door small, station wagons, sports/speciality, etc…), and represents the average loss per car per year.

# **8. Linear Regression**
---


__Linear regression__ is a _basic_ and _commonly_ used type of __predictive analysis__.  The overall idea of regression is to examine two things: 
- Does a set of __predictor variables__ do a good job in predicting an __outcome__ (dependent) variable?  
- Which variables in particular are __significant predictors__ of the outcome variable, and in what way they do __impact__ the outcome variable?  

These regression estimates are used to explain the __relationship between one dependent variable and one or more independent variables__.  The simplest form of the regression equation with one dependent and one independent variable is defined by the formula :<br/>
$y = \beta_0 + \beta_1x$


Now, we can see how we can use Linear Regression to estimate the price of old cars.

<a name = Section81></a>
## 8.1. Linear Regression Assumptions:

1. The independent variables should not be correlated. Absence of this phenomenon is known as multicollinearity. 
2. There should be no correlation between the residual (error) terms. Absence of this phenomenon is known as Autocorrelation.
3. The error terms must have constant variance. This phenomenon is known as homoskedasticity. The presence of non-constant variance is referred to heteroskedasticity.
4. The Target Variable must be normally distributed
5. Linear relationship between Independent and Dependent Variable


Let us analyse each of the above assumptions and verify all the assumptions are satisfied by the dataset. 

**1. The independent variables should not be correlated**

```python
fig = plt.figure(figsize = (20,20))

sns.heatmap(dataset.corr(), annot = True)

plt.show()
```

```python
dataset.drop(['fuel-type','compression-ratio','wheel-base','engine-size','curb-weight','horsepower','highway-mpg'], axis = 1, inplace = True)
```

```python
fig = plt.figure(figsize = (20,20))

sns.heatmap(dataset.corr(), annot = True)

plt.show()
```

Assumption 1 is satisfied

**2.There should be no correlation between the residual (error) terms**

This can be done after we build the model

**3. The error terms must have constant variance.**

This can be a safe assumption

**4. The target variable should be normally distributed**

## **8.2. Data Preparation for ML**

```python
dataset.info()
```

**Observations:**:


- We can see certain features which have numerical data are of type Object. In order to apply Regression, we need to encode these categorical features to numerical features. This is called as Encoding.

- The features to be encoded are **normailized-losses, make, fuel-type, aspiration, num-of-doors, body-style, drive-wheels, engine-location, engine-type, num-of-cylinders, fuel-sytem, bore, stroke, horsepower, peak-rpm.**



```python
# As the normalized-type feature is numeric, we are converting the type.

dataset['normalized-losses'] = dataset['normalized-losses'].astype(np.int64)

# Make 

dataset['make'].value_counts()

dataset.replace({'make':{'toyota':0,'nissan':1,'subaru':2,'mazda':3,'mitsubishi':4,'honda':5,'volkswagen':6,'dodge':7,'volvo':8,'peugot':9,'mercedes-benz':10,'plymouth':11,'bmw':12,'saab':13,'audi':14,'porsche':15,'jaguar':16,'chevrolet':17,'alfa-romero':18,'isuzu':19,'mercury':20}}, inplace = True)

# Fuel Type

#dataset['fuel-type'].value_counts()

#dataset.replace({'fuel-type':{'gas':0, 'diesel':1}}, inplace = True)


#  Aspiration

dataset['aspiration'].value_counts()

dataset.replace({'aspiration':{'std':0, 'turbo':1}}, inplace = True)

# Num of Doors

dataset['num-of-doors'].value_counts()

dataset.replace({'num-of-doors':{'four':0,'two':1}}, inplace = True)

# Body Style

dataset['body-style'].value_counts()

dataset.replace({'body-style':{'sedan':0,'hatchback':1,'wagon':2,'hardtop':3,'convertible':4}}, inplace = True)

# drive-wheels

dataset['drive-wheels'].value_counts()

dataset.replace({'drive-wheels':{'fwd':0,'rwd':1,'4wd':2}}, inplace = True)

# Engine-location

dataset['engine-location'].value_counts()

dataset.replace({'engine-location':{'front':0,'rear':1}}, inplace = True)

# Engine Type

dataset['engine-type'].value_counts()

dataset.replace({'engine-type':{'ohc':0,'ohcf':1,'ohcv':2,'dohc':3,'l':4}}, inplace = True)

#Num-of-cylinders

dataset['num-of-cylinders'].value_counts()

dataset.replace({'num-of-cylinders':{'four':0,'six':1,'five':2,'eight':3,'three':4,'twelve':5}}, inplace = True)

# Fuel System

dataset['fuel-system'].value_counts()

dataset.replace({'fuel-system':{'mpfi':0, '2bbl':1, 'idi':2, '1bbl':3,'spdi':4, 'mfi':5, 'spfi':6}}, inplace = True)

# Bore

dataset['bore'] = dataset['bore'].astype(float)


# Stroke

dataset['stroke'] = dataset['stroke'].astype(float)

# Horsepower

#dataset['horsepower'] = dataset['horsepower'].astype(np.int64)

# Peak RPM

dataset['peak-rpm'] = dataset['peak-rpm'].astype(np.int64)
```
